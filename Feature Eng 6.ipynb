### Question 3

Q3. Suppose you are conducting a study to examine the relationship between the number of hours of exercise per week and body mass index (BMI) in a sample of adults. You collected data on both variables for 50 participants. Calculate the Pearson correlation coefficient and the Spearman's rank correlation between these two variables and compare the results.

__Answer__


```python
import numpy as np
import pandas as pd

# Generate random data for exercise hours and BMI
exercise_hours = np.random.randint(1, 20, size=50)
bmi = np.random.uniform(18, 35, size=50)

df = pd.DataFrame({"exercise_hours":exercise_hours, "bmi":bmi})
df

#Analysis

pearson_result = df.corr(method = "pearson")
spearman_result = df.corr(method = "spearman")

print("This is pearman's correlation analysis result:\n {}".format(pearson_result))
print("\n\nThis is spearman's correlation analysis result:\n {}".format(spearman_result))



```

```python
### Another way to do this!!!!
import numpy as np
from scipy.stats import pearsonr, spearmanr

# Generate random data for exercise hours and BMI
exercise_hours = np.random.randint(1, 20, size=50)
bmi = np.random.uniform(18, 35, size=50)

# Calculate Pearson correlation coefficient
pearson_corr, _ = pearsonr(exercise_hours, bmi)
print("Pearson correlation coefficient:", pearson_corr)

# Calculate Spearman's rank correlation
spearman_corr, _ = spearmanr(exercise_hours, bmi)
print("Spearman's rank correlation:", spearman_corr)
```

Interpretation:

Both results are almost the same showing that there is a weak positive correlation between exercise hours and BMI . Closely oberving the result of pearson correlation coefficient and spearman correlation coefficient there is not much difference in the result. In the case of this analysis, it simply indicate that this data does not have outliers that could have greatly impacted
pearson correlation analysis result.

### Question 1

Q6. A company is interested in examining the relationship between the number of sales calls made per day and the number of sales made per week. The company collected data on both variables from a sample of 30 sales representatives. Calculate the Pearson correlation coefficient between these two variables.

__Answer__

```python
import numpy as np
import pandas as pd

# Set the random seed for reproducibility
np.random.seed(42)

# Generate random data for the number of sales calls and number of sales
sales_calls = np.random.randint(low=50, high=200, size=30)
sales_made = np.random.randint(low=5, high=25, size=30)

# Create a pandas DataFrame to store the data
data = pd.DataFrame({'Sales_calls': sales_calls, 'Sales_made': sales_made})

# Calculate the Pearson correlation coefficient
corr_coeff = data.corr(method = "pearson")

# Display the correlation coefficient
print("Pearson correlation coefficient:\n", corr_coeff)
```

Interpretation:

There is a moderate positive relationship between sale call and sale made base on the result of this analysis as the pearson coefficient is 0.344524. This indicates that with increasing number of sales call there will be increase in sale made too!!
